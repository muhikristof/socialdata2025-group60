```python
# Filter for DUI cases
dui_df = df[df["Category"] == "DUI"].copy()

# Extract the hour from the Time column
dui_df["Hour"] = pd.to_datetime(dui_df["Time"], format="%H:%M", errors="coerce").dt.hour
dui_df["WeekHour"] = dui_df["Hour"] + pd.to_datetime(dui_df["Date"]).dt.dayofweek * 24

# Count the number of DUI incidents per hour
hourly_counts = dui_df["Hour"].value_counts().sort_index()
weekly_counts = dui_df["WeekHour"].value_counts().sort_index()

# Ensure all 24 hours are represented in the data
theta = np.linspace(0, 2 * np.pi, 24, endpoint=False)  # Angle positions for 24 hours
radii = hourly_counts.reindex(range(24), fill_value=0)  # Fill missing hours with zero

# Create the polar bar chart for 24-hour data
fig, ax = plt.subplots(figsize=(5, 5), subplot_kw={'projection': 'polar'})
bars = ax.bar(theta, radii, width=2 * np.pi / 24, bottom=0, color='red', alpha=0.7)

# Set labels and title
ax.set_xticks(theta)
ax.set_xticklabels([f"{h}:00" for h in range(24)], fontname="Arial")
ax.set_title("DUI Incidents Over 24 Hours", va='bottom', fontname="Arial")

# Show the plot
plt.show()

# Create the polar bar chart for 168-hour (weekly) data
theta_week = np.linspace(0, 2 * np.pi, 168, endpoint=False)  # Angle positions for 168 hours
radii_week = weekly_counts.reindex(range(168), fill_value=0)  # Fill missing hours with zero

daylabels = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
fig, ax_week = plt.subplots(figsize=(5, 5), subplot_kw={'projection': 'polar'})
bars_week = ax_week.bar(theta_week, radii_week, width=2 * np.pi / 168, bottom=0, color='blue', alpha=0.7)

# Set labels and title
ax_week.set_xticks(theta_week[::24])
ax_week.set_xticklabels(daylabels, fontname="Arial")
ax_week.set_title("DUI Incidents Over 168 Hours (Week)", va='bottom', fontname="Arial")

# Show the plot
plt.show()
```

> These two polar bar charts visualize the distribution of DUI (driving under influence) incidents over different time periods.

> DUI Incidents Over 24 Hours:
The majority of DUI incidents occur late at night, peaking between midnight and 2 AM. The frequency drops sharply in the early morning hours and remains low during the daytime. This pattern aligns with typical social behavior, where alcohol consumption tends to happen during late-night activities, followed by individuals attempting to drive afterward, often resulting in DUI offenses.

> DUI Incidents Over a Week:
The highest concentration of DUI incidents occurs on weekends, particularly late Saturday night into early Sunday morning. This suggests that DUI offenses are more likely to occur during leisure times associated with weekend activities and gatherings. Weekdays show relatively lower DUI numbers, reinforcing the link between social drinking patterns and weekend behavior, where the likelihood of drinking and driving increases.

```python
crime_combined = df_noDuplicates.copy()

# Ensure Date is in datetime format and Time is a string
# Ensure Date is in datetime format
crime_combined["Date"] = pd.to_datetime(crime_combined["Date"], errors="coerce")

# Ensure Time is properly formatted to HH:MM:SS (handle missing seconds if needed)
crime_combined["Time"] = crime_combined["Time"].str.strip()  # Remove spaces

# Fix cases where Time might be in HH:MM format by adding seconds
crime_combined["Time"] = crime_combined["Time"].apply(lambda x: x + ":00" if len(x) == 5 else x)

# Convert Time to timedelta
crime_combined["Datetime"] = crime_combined["Date"] + pd.to_timedelta(crime_combined["Time"], errors="coerce")

# Combine Date and Time correctly
crime_combined["Datetime"] = crime_combined["Date"] + pd.to_timedelta(crime_combined["Time"])
```

```python
# Select a specific month (e.g., January 2015)
selected_month = "2015-01"
filtered_crimes = crime_combined[crime_combined["Datetime"].dt.to_period("M") == selected_month]

# Filter data for Burglary and Drug Violation
burglary_data = filtered_crimes[filtered_crimes["Category"] == "Burglary"]["Latitude"].dropna()
drug_violation_data = filtered_crimes[filtered_crimes["Category"] == "Drug Violation"]["Latitude"].dropna()

# --- Probability Plots ---
plt.figure(figsize=(12, 6))

# Q-Q Plot for Burglary
plt.subplot(1, 2, 1)
stats.probplot(burglary_data, dist="norm", plot=plt)
plt.xlabel("Normal Probits")
plt.ylabel("Latitude")
plt.title("Q-Q Plot of Burglary (Latitude)")

# Q-Q Plot for Drug Violation
plt.subplot(1, 2, 2)
stats.probplot(drug_violation_data, dist="norm", plot=plt)
plt.xlabel("Normal Probits")
plt.ylabel("Latitude")
plt.title("Q-Q Plot of Drug Violation (Latitude)")

# Show plots
plt.tight_layout()
plt.show()
```

> The Q-Q plots compare the latitude distributions of burglary and drug violation incidents against a normal distribution. The burglary data (left) follows the normality line relatively well, with minor deviations at the extremes, suggesting an approximately normal distribution. In contrast, the drug violation data (right) shows more pronounced deviations, particularly at the tails, indicating a non-normal distribution.

```python
# List of focus crime categories
focus_crimes = [
    "Weapons Offense", "Prostitution", "Robbery", "Burglary", "Assault",
    "Liquor Laws", "Drug Violation", "Trespassing", "Larceny Theft",
    "Vandalism", "Vehicle Theft", "Stolen Property"
]

# Count number of crimes per day per category
daily_crime_counts = crime_combined[crime_combined["Category"].isin(focus_crimes)]
daily_crime_counts = daily_crime_counts.groupby(["Date", "Category"]).size().reset_index(name="Crime_Count")

# Pivot to get crime categories as columns
crime_pivot = daily_crime_counts.pivot(index="Date", columns="Category", values="Crime_Count").fillna(0)

# --- Box Plot ---
plt.figure(figsize=(14, 6))

# Create horizontal box plot for crime counts per category
plt.boxplot([crime_pivot[col] for col in focus_crimes], tick_labels=focus_crimes, vert=False, patch_artist=True)

# Formatting
plt.yticks(fontsize=10)  # Adjust font size for readability
plt.xlabel("Daily Crime Count", fontsize=12)
plt.title("Daily Crime Count Distribution for Focus Crimes", fontsize=14)
plt.grid(axis="x", linestyle="--", alpha=0.6)  # Adjust grid to match horizontal layout

# Show plot
plt.tight_layout()
plt.show()
```

> This visualization highlights the spread, median, and outliers for each crime type in a compact way. Some days, Larceny Theft incidents spike far beyond the typical range.

> - Some crimes (e.g., Weapons Offense, Prostitution) likely have a narrower interquartile range (IQR)—meaning they are more consistent day-to-day.
> - Others (like Larceny Theft) have wider boxes and more outliers, indicating more variability.

```python
# Extract hour and minute, then convert to decimal hours
crime_combined["Hour"] = crime_combined["Datetime"].dt.hour
crime_combined["Minute"] = crime_combined["Datetime"].dt.minute
crime_combined["Time_Decimal"] = crime_combined["Hour"] + (crime_combined["Minute"] / 60)

# Filter dataset for focus crimes
filtered_crimes = crime_combined[crime_combined["Category"].isin(focus_crimes)]

# Ensure data exists
filtered_crimes = filtered_crimes.dropna(subset=["Time_Decimal"])

# Remove empty categories
filtered_focus_crimes = [crime for crime in focus_crimes if not filtered_crimes[filtered_crimes["Category"] == crime]["Time_Decimal"].empty]

plt.figure(figsize=(14, 6))

plt.boxplot(
    [filtered_crimes[filtered_crimes["Category"] == crime]["Time_Decimal"] for crime in filtered_focus_crimes], 
    tick_labels=filtered_focus_crimes, vert=True, patch_artist=True
)

# Formatting
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.ylabel("Time of Day (Decimal Hours)", fontsize=12)
plt.title("Time of Day Distribution for Focus Crimes", fontsize=14)
plt.ylim(0, 24)
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```
